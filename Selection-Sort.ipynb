    <h1>Selection Sort</h1>
    <p>
        Selection Sort is a simple sorting algorithm that works by repeatedly finding the minimum element from the unsorted portion of the array and swapping it with the element at the beginning of the unsorted portion. This process continues until the entire array is sorted.
    </p>
    <h2>How Selection Sort Works</h2>
    <p>
        The algorithm works as follows:
    </p>
    <ol>
        <li>
            <strong>Find the minimum element:</strong>
            <p>
                Start with the first element of the array and assume it is the smallest. Traverse the unsorted portion of the array to find the actual smallest element.
            </p>
        </li>
        <li>
            <strong>Swap with the first unsorted element:</strong>
            <p>
                Once the smallest element is found, swap it with the first element of the unsorted portion.
            </p>
        </li>
        <li>
            <strong>Move the boundary of the sorted portion:</strong>
            <p>
                After each swap, the sorted portion of the array grows by one element, and the unsorted portion shrinks by one element.
            </p>
        </li>
        <li>
            <strong>Repeat until the array is sorted:</strong>
            <p>
                Repeat the process for the remaining unsorted portion of the array until the entire array is sorted.
            </p>
        </li>
    </ol>
    <h2>Time Complexity</h2>
    <ul>
        <li><strong>Best Case:</strong> O(n²) (even if the array is already sorted).</li>
        <li><strong>Average Case:</strong> O(n²).</li>
        <li><strong>Worst Case:</strong> O(n²).</li>
    </ul>
    <h2>Space Complexity</h2>
    <p>
        Selection Sort is an <strong>in-place</strong> sorting algorithm, meaning it requires only O(1) additional space.
    </p>
    <h2>Advantages</h2>
    <ul>
        <li>Simple to understand and implement.</li>
        <li>No additional memory is required (in-place sorting).</li>
        <li>Performs well for small datasets.</li>
    </ul>
    <h2>Limitations</h2>
    <ul>
        <li>Inefficient for large datasets due to its O(n²) time complexity.</li>
        <li>Not stable (does not preserve the relative order of equal elements).</li>
    </ul>
    <h2>When to Use Selection Sort</h2>
    <ul>
        <li>When the array is small.</li>
        <li>When memory usage is a concern (since it is in-place).</li>
        <li>When simplicity is more important than performance.</li>
    </ul>

```python
import random

# Number of random numbers
n = 10
start = 1
end = 100

# Creating the list of random numbers
Random_list = [random.randint(start, end) for _ in range(n)]
print("\nOriginal List:", Random_list, "\n")

def selection_sort(A):
   
    n = len(A)
    for i in range(n - 1):
        index = i  
        for j in range(i + 1, n):
            if A[index] > A[j]:  
                index = j
        A[i], A[index] = A[index], A[i]
    return A


# Example
sorted_list = selection_sort(Random_list)
print("Sorted Array:", sorted_list)
```
