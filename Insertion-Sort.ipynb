```python
import random

# Number of random numbers
n = 10
# The interval of random numbers
start = 1
end = 100

# Creating the list of random numbers
Random_list = [random.randint(start, end) for _ in range(n)]
print("\nOriginal List:", Random_list, "\n")

def insertion_sort(array):
    
    for k in range(1, len(array)):  
        item = array[k]  
        i = k  
        while i > 0 and array[i - 1] > item:
            array[i] = array[i - 1]  
            i -= 1
        array[i] = item 
    return array

# Example
sorted_list = insertion_sort(Random_list)
print("Sorted Array:", sorted_list)
```

    <h1>Insertion Sort</h1>
    <p>
        Insertion Sort is a simple and intuitive sorting algorithm that builds the final sorted array one element at a time. It works similarly to how you might sort a hand of playing cards: you pick one card at a time and insert it into its correct position among the already sorted cards.
    </p>
    <h2> How Insertion Sort Works</h2>
    <p>
        The algorithm works by dividing the array into two parts:
    </p>
    <ul>
        <li><strong>Sorted Part:</strong> Initially, this part contains only the first element of the array.</li>
        <li><strong>Unsorted Part:</strong> This part contains the remaining elements of the array.</li>
    </ul>
    <p>
        The algorithm then iterates over the unsorted part, taking one element at a time and inserting it into its correct position in the sorted part.
    </p>
    <h2>Steps of Insertion Sort</h2>
    <ol>
        <li>
            <strong>Start with the second element:</strong>
            <p>
                Assume the first element is already sorted. Begin with the second element (index 1) and consider it as the key to be inserted into the sorted part.
            </p>
        </li>
        <li>
            <strong>Compare and Shift:</strong>
            <p>
                Compare the key with the elements in the sorted part (from right to left). If the key is smaller than an element, shift that element one position to the right.
            </p>
        </li>
        <li>
            <strong>Insert the Key:</strong>
            <p>
                Once the correct position is found, insert the key into its place.
            </p>
        </li>
        <li>
            <strong>Repeat:</strong>
            <p>
                Repeat the process for all elements in the unsorted part until the entire array is sorted.
            </p>
        </li>
    </ol>
    <h2>Time Complexity</h2>
    <ul>
        <li><strong>Best Case:</strong> O(n) (when the array is already sorted).</li>
        <li><strong>Average Case:</strong> O(n²).</li>
        <li><strong>Worst Case:</strong> O(n²) (when the array is sorted in reverse order).</li>
    </ul>
    <h2>Space Complexity</h2>
    <p>
        Insertion Sort is an <strong>in-place</strong> sorting algorithm, meaning it requires only O(1) additional space.
    </p>
    <h2>Advantages</h2>
    <ul>
        <li>Simple to understand and implement.</li>
        <li>Efficient for small datasets.</li>
        <li>Stable (preserves the relative order of equal elements).</li>
        <li>Works well with nearly sorted arrays.</li>
    </ul>
    <h2>Limitations</h2>
    <ul>
        <li>Inefficient for large datasets due to its O(n²) time complexity.</li>
        <li>Performs poorly on reverse-ordered arrays.</li>
    </ul>
