```python
import random

# Number of random numbers
n = 10
start = 1
end = 100

# Creating the list of random numbers
Random_list = [random.randint(start, end) for _ in range(n)]
print("\nOriginal List:", Random_list, "\n")

def bubble_sort(array):
   
    n = len(array)
    for i in range(n - 1):
        for j in range(n - 1, i, -1):
            if array[j] < array[j - 1]:
                array[j], array[j - 1] = array[j - 1], array[j] 
        print(f"After pass {i + 1}: {array}")  
    return array


# Example
sorted_list = bubble_sort(Random_list)
print("\nFinal Sorted Array:", sorted_list)
```

    <h1>Bubble Sort</h1>
    <p>
        Bubble Sort is a simple sorting algorithm that repeatedly steps through the list, compares adjacent elements, and swaps them if they are in the wrong order. This process is repeated until the list is sorted.
    </p>
    <h2>How Bubble Sort Works</h2>
    <p>
        The algorithm works as follows:
    </p>
    <ol>
        <li>
            <strong>Start at the beginning of the array:</strong>
            <p>
                Compare the first two elements. If the first element is greater than the second, swap them.
            </p>
        </li>
        <li>
            <strong>Move to the next pair:</strong>
            <p>
                Continue comparing and swapping adjacent elements until you reach the end of the array.
            </p>
        </li>
        <li>
            <strong>Repeat the process:</strong>
            <p>
                After each pass, the largest unsorted element "bubbles up" to its correct position at the end of the array. Repeat the process for the remaining unsorted portion of the array.
            </p>
        </li>
        <li>
            <strong>Stop when no swaps are needed:</strong>
            <p>
                If no swaps are made during a pass, the array is already sorted, and the algorithm can terminate early.
            </p>
        </li>
    </ol>
    <h2>Time Complexity</h2>
    <ul>
        <li><strong>Best Case:</strong> O(n) (when the array is already sorted).</li>
        <li><strong>Average Case:</strong> O(n²).</li>
        <li><strong>Worst Case:</strong> O(n²) (when the array is sorted in reverse order).</li>
    </ul>
    <h2>Space Complexity</h2>
    <p>
        Bubble Sort is an <strong>in-place</strong> sorting algorithm, meaning it requires only O(1) additional space.
    </p>
    <h2>Advantages</h2>
    <ul>
        <li>Simple to understand and implement.</li>
        <li>No additional memory is required (in-place sorting).</li>
        <li>Stable (preserves the relative order of equal elements).</li>
    </ul>
    <h2>Limitations</h2>
    <ul>
        <li>Inefficient for large datasets due to its O(n²) time complexity.</li>
        <li>Performs poorly on reverse-ordered arrays.</li>
    </ul>
    <h2>When to Use Bubble Sort</h2>
    <ul>
        <li>When the array is small or nearly sorted.</li>
        <li>When simplicity is more important than performance.</li>
        <li>When memory usage is a concern (since it is in-place).</li>
    </ul>
